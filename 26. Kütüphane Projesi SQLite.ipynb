Kütüphane Uygulaması

```python
import sqlite3 
con = sqlite3.connect("Kütüphane.db") 
cursor = con.cursor() 
cursor.execute("CREATE TABLE IF NOT EXISTS kitaplık (İsim TEXT, Katagori TEXT, Yazar TEXT, Yayınevi TEXT, Sayfa_Sayısı INT)")
con.commit() 

def kitap_ekle():
    isim=input("Kitabın ismini giriniz:")
    katagori=input("Kitabın katagorisini giriniz:")
    yazar=input("Kitabın yazar ismini giriniz:")
    yayın=input("Kitabın yayın evini giriniz:")
    sayfa=int(input("Kitabın sayfa sayısını giriniz:"))
    cursor.execute("Insert into kitaplık Values(?,?,?,?,?)",(isim,katagori,yazar,yayın,sayfa))
    con.commit()
    print(f"{isim} isimli kitap eklendi.")

def kitap_güncelle():
    eski_isim=input("Eski kitap ismini giriniz:")
    yeni_isim=input("Yeni kitap ismini giriniz:")
    cursor.execute("UPDATE kitaplık SET İsim = ? WHERE İsim =  ?",(yeni_isim,eski_isim,))
    con.commit()
    print(f"{eski_isim} isimli kitap {yeni_isim} olarak güncellendi.")

def kitap_sil():
    kitap=input("Silmek istediğiniz kitabın ismini giriniz:")
    cursor.execute("Delete From kitaplık where İsim=?",(kitap,))
    con.commit()
    print(f"{kitap} isimli kitap silindi.")

def envanter():
    cursor.execute("Select * From kitaplık")
    liste=cursor.fetchall()
    print("Kütüphanemizde mevcut kitaplar\n{}".format("-"*30))
    for i in liste:
        print(i)

def kitap_sorgula():
    isim=input("Sorgulamak istediğiniz kitapın isimini giriniz:")
    cursor.execute("Select * From kitaplık where İsim=?",(isim,)) 
    liste=cursor.fetchall()
    print("Aradığınız kitabın bilgileri\n{}".format("-"*30))
    for i in liste:
        print(i)

def katagori_filtre():
    katagori=input("Fitrelemek istediğiniz katagoriyi giriniz:")
    cursor.execute("Select * From kitaplık where Katagori=?",(katagori,)) 
    liste=cursor.fetchall()
    print(f"{katagori} katagorisindeki kitaplar aşağıda sıralanmıştır")
    for i in liste:
        print(i)
        
def yayın_filtre():
    yayın=input("Filtrelemek istediğiniz yayınevini giriniz:")
    cursor.execute("Select * From kitaplık where Yayınevi=?",(yayın,)) 
    liste=cursor.fetchall()
    print("Filtrelediğiniz yayının kitapları aşağıda sıralanmıştır")
    for i in liste:
        print(i)

print("Kütüphane Programı \n{}".format("-"*18))
print("1-Kitap ekle \n2-Kitap güncelle \n3-Kitap sil \n4-Kitaplık envanteri \n5-Kitap sorgula \n6-Katagori filtrele \n7-Yayınevi filtrele \n8-Çıkış('q')")
print("{}".format("-"*18))
while True:
    seçenek=input("İşlem seçiniz:")
    seçenek.lower()
    if seçenek=="1" or seçenek=="kitap ekle":
        kitap_ekle()
    elif seçenek=="2" or seçenek=="kitap güncelle":
        kitap_güncelle()
    elif seçenek=="3" or seçenek=="kitap sil":
        kitap_sil()
    elif seçenek=="4" or seçenek=="kitaplık envanteri":
        envanter()
    elif seçenek=="5" or seçenek=="kitap sorgula":
        kitap_sorgula()
    elif seçenek=="6" or seçenek=="katagori filtrele":
        katagori_filtre()
    elif seçenek=="7" or seçenek=="yayınevi filtrele":
        yayın_filtre()
    elif seçenek=="8" or seçenek=="q" or seçenek=="çıkış":
        print("Çıkış yapılıyor...")
        break
    else:
        print("Geçersiz işlem")
```
