Döviz Programı

```python
print("Döviz Programına Hoşgeldiniz...")
print("-"*33)
import requests
from bs4 import BeautifulSoup
url = "https://www.doviz.com/"
response=requests.get(url)
html_iceriği=response.content
soup=BeautifulSoup(html_iceriği,"html.parser")

isim = soup.find_all("span",{"class":"name"})
değer = soup.find_all("span",{"class":"value"})

altın=değer[0].text
altın=float(altın.replace('.', '').replace(',', '.'))
dolar=değer[1].text
dolar=float(dolar.replace(',', '.'))
euro=değer[2].text
euro=float(euro.replace(',', '.'))
sterlin=değer[3].text
sterlin=float(sterlin.replace(',', '.'))
bıst=değer[4].text
bıst=float(bıst.replace('.', '').replace(',', '.'))
btc=değer[5].text
btc=float(btc.replace(',', '.').lstrip("$"))
btc=round(btc*dolar,2)
gümüş=değer[6].text
gümüş=float(gümüş.replace(',', '.'))
brent=değer[7].text
brent=float(brent.replace(',', '.').lstrip("$"))
brent=round(brent*dolar,2)

veriler=[
   ("1-Gram altın",altın,"₺  (1)"),
   ("2-Dolar",dolar,"₺  (2)"),
   ("3-Euro",euro,"₺  (3)"),
   ("4-Sterlin",sterlin,"₺  (4)"),
   ("5-Bist",bıst,"₺  (5)"),
   ("6-BTC",btc,"₺  (6)"),
   ("7-Gümüş",gümüş,"₺  (7)"),
   ("8-Brent",brent,"₺  (8)")
]

for veri in veriler:
    print(f"{veri[0]:<15} {veri[1]:>10.2f} {veri[2]:<5}")

print("-"*33)
a=input("Birinci para birimini giriniz:")
a.lower()
if a=="altın" or a=="1":
   a=altın
   print("Altın")
elif a=="dolar" or a=="2":
   a=dolar
   print("Dolar")
elif a=="euro" or a=="3":
   a=euro
   print("Euro")
elif a=="sterlin" or a=="4":
   a=sterlin
   print("Sterlin")
elif a=="bıst" or a=="5":
   a=bıst
   print("Bist")
elif a=="btc" or a=="6":
   a=btc
   print("BTC")
elif a=="gümüş" or a=="7":
   a=gümüş
   print("Gümüş")
elif a=="brent" or a=="8":
   a=brent
   print("Brent")
else:
   print("Geçerli bir değer giriniz!")

b=int(input("Miktar giriniz:"))
print(b,"Adet")
tot = a*b
print("-"*33)
print("Bakiyeniz:",round(tot,3),"₺")
print("-"*33)

c=input("İkinci para birimini giriniz:")
c.lower()
if c=="altın" or c=="1":
   c=altın
   print("Altın")
elif c=="dolar" or c=="2":
   c=dolar
   print("Dolar")
elif c=="euro" or c=="3":
   c=euro
   print("Euro")
elif c=="sterlin" or c=="4":
   c=sterlin
   print("Sterlin")
elif c=="bıst" or c=="5":
   c=bıst
   print("Bist")
elif c=="btc" or c=="6":
   c=btc
   print("BTC")
elif c=="gümüş" or c=="7":
   c=gümüş
   print("Gümüş")
elif c=="brent" or c=="8":
   c=brent
   print("Brent")
else:
   print("Geçerli bir değer giriniz!")

print(round(tot/c,3))
print("-"*33)
```
