TV Kumandası Uygulaması

```python
import time
class kumanda:
    def __init__(self,tv_durumu="Kapalı",tv_ses=10,kanallar=["TRT","TLC","D-MAX"],kanal="TRT"):
        self.tv_durumu=tv_durumu
        self.tv_ses=tv_ses
        self.kanallar=kanallar
        self.kanal=kanal

    def aç(self):
        if self.tv_durumu=="Açık":
            print("TV zaten açık")
        else:
            self.tv_durumu="Açık"
            print("Tv açılıyor...")
            time.sleep(2)
            print("TV açıldı")
    def kapat(self):
        if self.tv_durumu=="Kapalı":
            print("TV zaten kapalı")
        else:
            self.tv_durumu="Kapalı"
            print("TV kapatılıyor...")
            time.sleep(2)
            print("TV kapandı")
        
    def sesartı(self):
        if self.tv_durumu=="Açık":
            if self.tv_ses>=100:
                self.tv_ses=self.tv_ses
                print("Ses Düzeyi:",self.tv_ses)
            else:
                self.tv_ses+=1
                print("Ses Düzeyi:",self.tv_ses)
        elif self.tv_durumu=="Kapalı":
            print("TV kapalı açmak için 1'i tuşlayınız.")
    def seseksi(self):
        if self.tv_durumu=="Açık":
            if self.tv_ses==0:
                self.tv_ses=self.tv_ses
                print("Ses Düzeyi:",self.tv_ses)
            else:
                self.tv_ses-=1
                print("Ses Düzeyi:",self.tv_ses)
        elif self.tv_durumu=="Kapalı":
            print("TV kapalı açmak için 1'i tuşlayınız.")
    def kanalartı(self):
        if self.tv_durumu=="Açık":
            a=self.kanallar.index(self.kanal)
            if a<(len(self.kanallar)-1):
                self.kanal=self.kanallar[a+1]
                print(self.kanal)
            else:
                self.kanal=self.kanallar[(len(self.kanallar)-(a+1))]
                print(self.kanal) 
        elif self.tv_durumu=="Kapalı":
            print("TV kapalı açmak için 1'i tuşlayınız.")      
    def kanaleksi(self):
        if self.tv_durumu=="Açık":
            a=self.kanallar.index(self.kanal)
            self.kanal=self.kanallar[a-1]
            print(self.kanal)
        elif self.tv_durumu=="Kapalı":
            print("TV kapalı açmak için 1'i tuşlayınız.")
    def kanalekle(self):
        if self.tv_durumu=="Açık":
            self.kanallar.append(input("Eklemek istediğiniz kanal:"))
            print("Kanal ekleniyor...")
            time.sleep(2)
            new=self.kanallar[len(self.kanallar)-1]
            print(f"{new} kanalı eklendi")
        elif self.tv_durumu=="Kapalı":
            print("TV kapalı açmak için 1'i tuşlayınız.")
    def kanalsil(self):
        if self.tv_durumu=="Açık":
            delete=input("Silmek istediğiniz kanal:")
            self.kanallar.remove(delete)
            print("Kanal siliniyor...")
            time.sleep(2)
            print(f"{delete} kanalı silindi")
        elif self.tv_durumu=="Kapalı":
            print("TV kapalı açmak için 1'i tuşlayınız.")
    def kanallist(self):
        if self.tv_durumu=="Açık":    
            print(self.kanallar)
        elif self.tv_durumu=="Kapalı":
            print("TV kapalı açmak için 1'i tuşlayınız.")
kumanda=kumanda()

print("""
Tv kumandası uygulaması:
************************
1- TV aç
2- TV kapat
3- Ses artır
4- Ses azalt
5- Kanal değiştir +
6- Kanal değiştir -
7- Kanal ekle
8- Kanal sil
9- Kanal listesi
q- Çıkış
************************                  
""")
while True:
    işlem=input("İşlem giriniz:")
    if işlem=="1":
        kumanda.aç()
    elif işlem=="2":
        kumanda.kapat()
    elif işlem=="3":
        kumanda.sesartı()
    elif işlem=="4":
        kumanda.seseksi()
    elif işlem=="5":
        kumanda.kanalartı()
    elif işlem=="6":
        kumanda.kanaleksi()
    elif işlem=="7":
        kumanda.kanalekle()
    elif işlem=="8":
        kumanda.kanalsil()
    elif işlem=="9":
        kumanda.kanallist()
    elif işlem=="q"or"Q":
        print("Sistemden çıkıldı")
        break
    else:
        print("Geçerli bir işlem giriniz")
```
