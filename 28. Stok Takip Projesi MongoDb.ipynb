Stok Takip Uygulaması

```python
import pymongo
myclient=pymongo.MongoClient("mongodb://localhost:27017/")
mydb=myclient["node-app"]
mycollection=mydb["ürünler"]

def ürün_ekle():
    ad = input("Ürün adı: ")
    miktar = int(input("Stok miktarı: "))
    fiyat = float(input("Ürün fiyatı: "))
    ürün = {"ad": ad,"miktar": miktar,"fiyat": fiyat}
    mycollection.insert_one(ürün)
    print("Ürün başarıyla eklendi!")

def envanter():
    for ürün in mycollection.find():
        print(f"Ürün adı: {ürün['ad']}, Stok miktarı: {ürün['miktar']}, Fiyat: {ürün['fiyat']}")

def ürün_sil():
    ad = input("Silmek istediğiniz ürünün adını girin: ")
    mycollection.delete_one({"ad": ad})
    print(f"{ad} ürünü başarıyla silindi!")

def ürün_güncelle():
    ad = input("Güncellemek istediğiniz ürünün adını girin: ")
    yeni_ad = input("Yeni ürün adı (değiştirmek istemiyorsanız boş bırakın): ")
    yeni_miktar = input("Yeni stok miktarı (değiştirmek istemiyorsanız boş bırakın): ")
    yeni_fiyat = input("Yeni ürün fiyatı (değiştirmek istemiyorsanız boş bırakın): ")

    güncelleme = {}
    if yeni_ad:
        güncelleme["ad"] = yeni_ad
    if yeni_miktar:
        güncelleme["miktar"] = int(yeni_miktar)
    if yeni_fiyat:
        güncelleme["fiyat"] = float(yeni_fiyat)

    mycollection.update_one({"ad": ad}, {"$set": güncelleme})
    print(f"{ad} ürünü başarıyla güncellendi!")

def ürün_sorgula():
    ad = input("Filtrelemek istediğiniz ürün adını girin: ")
    sonuç = mycollection.find({"ad": ad})
    for ürün in sonuç:
        print(f"Ürün adı: {ürün['ad']}, Stok miktarı: {ürün['miktar']}, Fiyat: {ürün['fiyat']}")

def ürün_fiyat_filtrele():
    en_az=int(input("En az değeri giriniz:"))
    en_fazla=int(input("En fazla değeri giriniz:"))
    result=mycollection.find({"$and":[{"fiyat":{"$gte":en_az}},{"fiyat":{"$lte":en_fazla}}]})
    for i in result:
        print(i)

def ürünleri_sırala():
    değer=input("1- Artan 2- Azalan ")
    değer.lower()
    if değer=="1" or "artan":
        result=mycollection.find().sort("fiyat",1)
        for i in result:
            print(i)
    elif değer=="2" or "azalan":
        result=mycollection.find().sort("fiyat",-1)
        for i in result:
            print(i)

while True:
    print("{} \nStok Takip Uygulaması \n{} \n1. Ürün Ekle \n2. Ürünleri Listele \n3. Ürün Sil \n4. Ürün Güncelle \n5. Ürün Sorgula \n6. Ürün Filtrele \n7. Ürünleri Sırala \n8. Çıkış ('q') \n{}".format(("-"*22),("-"*22),("-"*22)))
    secim = input("Seçiminizi yapın: ")
    if secim == '1':
        ürün_ekle()
    elif secim == '2':
        envanter()
    elif secim == '3':
        ürün_sil()
    elif secim == '4':
        ürün_güncelle()
    elif secim == '5':
        ürün_sorgula()
    elif secim == '6':
        ürün_fiyat_filtrele()
    elif secim == '7':
        ürünleri_sırala()
    elif secim == '8' or "q" or "Q":
        print("Çıkış yapılıyor..")
        break
    else:
        print("Geçersiz seçim!")
```
